```python
plt.plot([1,2,3], [1,4,9])
plt.plot([1,2,3], [1,8,16])
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('First Chart!')
plt.legend(['Series 1', 'Series 2'])
plt.show()
```
