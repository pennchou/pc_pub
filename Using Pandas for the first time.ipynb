```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = {'year': [2008, 2012, 2016],
        'attendees': [112, 132, 729],
        'average age': [24, 43, 31]}
df = pd.DataFrame(data)
df
```

```python
type(df['year'])
```

```python
earlier_than_2013 = df['year'] < 2013
df[earlier_than_2013]
```

```python
plt.plot(df['year'], df['attendees'])
plt.plot(df['year'], df['average age'])
plt.legend(['attendees', 'average age'])
plt.show()
```

```python
#import countries data from csv

countries = pd.read_csv('countries.csv')
countries
```

```python
##compare US population with China population

# US population by year

us = countries.loc[countries['country'] == 'United States']
china = countries.loc[countries['country'] == 'China']

plt.plot(us['year'], us['population']/10**6)
plt.plot(china['year'], china['population']/10**6)
plt.xlabel('year')
plt.ylabel('Population (M)')
plt.legend(['US', 'China'])
```

```python
## compare US and China population Growth

#us.iloc[1].population

plt.plot(us['year'], us['population']/us.iloc[0].population)
plt.plot(china['year'], china['population']/china.iloc[0].population)
plt.xlabel('year')
plt.ylabel('Population (M)')
plt.legend(['US', 'China'])
```

```python
us['growth%'] = us['population']/us.iloc[0].population

china['growth%'] = china['population']/china.iloc[0].population

plt.plot(us['year'], us['growth%'])
plt.plot(china['year'], china['growth%'])
plt.xlabel('year')
plt.ylabel('Population growth %')
plt.legend(['US', 'China'])
plt.show()


```
